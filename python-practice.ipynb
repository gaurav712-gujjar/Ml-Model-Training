```python
import numpy as np

#ARRAY CREATION
#Array creation 
#1D array
arr=np.array([1,2,3,4,5])
print(arr)
print("=============================")

#2D array
arr1=np.array([[1,2,3,4],[5,6,7,8]])
print(arr1)
print("=============================")

#Creates an array filled with zeros
arr=np.zeros((2,3))
print(arr)
print("=============================")

#Creates an array filled with ones
arr1=np.ones((3,3))
print(arr1)
print("=============================")

arr2=np.empty((2,3))
print(arr2)
print("=============================")

#print a range 
arr=np.arange(0,10,2)
print(arr)

#Creates evenly spaced values between two numbers
arr1=np.linspace(1,20,5)
print(arr1)
print("=============================")

#Creates an identity matrix
arr3=np.eye(3)
print(arr3)
print("=============================")


#Creates an array filled with a given value
arr4=np.full((3,3),7)
print(arr4)



```
